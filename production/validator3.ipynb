```python
import pandas as pd
from helper import get_files_from_folder
import os
import re
```

```python
MIN_SIZE = 512
BASE_DIR = "."
FOLDERS = [os.path.join(BASE_DIR, o) for o in os.listdir(BASE_DIR) if os.path.isdir(os.path.join(BASE_DIR,o))]
FOLDERS.sort()
FOLDERS = FOLDERS[1:-1]
```

```python
HEADER_EXP = re.compile("((Datum)|(Date));((Time of day|Uhrzeit));((Generation_DE (\w*(\s?)\&?)*\[MW\]);?)*$")
LINE_STR = "(\d\d.\d\d.\d{4});(\d\d:\d\d);((\d+;?)|(\d.\d\d\d;?)|(-;)|-|;)*$"
LINE_EXP = re.compile("(\d\d.\d\d.\d{4});(\d\d:\d\d);(\d+|(.\d\d\d)|-|;)*$")
```

```python
FILES_L = [get_files_from_folder(f) for f in FOLDERS]
FILES = [item for sublist in FILES_L for item in sublist]
```

```python
def get_lines(file):
    with open(file, 'r') as myfile:
        tmp = myfile.readlines()
        return tmp[0], tmp[1:]
```

```python
FILES = ['./2016_17/Weiher_201601010000_201712312345_21.csv']
```

```python
def get_m(count):
    match_string = LINE_STR.replace("*", "{" + str(count) + "}")
    return re.compile(match_string)
```

```python
result = {}
for file in FILES:
    #print(file)
    tmp = []
    valid = True
    try:
        header, lines = get_lines(file)
    except UnicodeDecodeError:
        print(file)
        pass
    h = HEADER_EXP.match(header[1:])
    if not h:
        valid = False
        tmp.append(-1)
    else:
        tl = len(h.string.split(";"))
        matcher = get_m(tl-2)
    for idx, line in enumerate(lines):
        x = matcher.match(line)
        if not x:
            tmp.append(idx)
            valid = False
    if not valid:
        result[file] = tmp
    
print(len(result))
```

```python
wrong_header = []
for k, v in result.items():
    #print(v)
    if -1 in v:
        wrong_header.append(k)

for h in wrong_header:
    result.pop(h) 
```

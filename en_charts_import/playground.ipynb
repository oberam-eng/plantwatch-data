```python
df = pd.DataFrame({'eexid' : [], 'year': [], 'value': []})
```

```python
eid = 0

for idx, obj in enumerate(pf):
    if isinstance(obj['eexUnitID'], list):
        eid = obj['eexUnitID'][0]
    else:
        eid = obj['eexUnitID']
    year, value = obj['values'][0]
    df.loc[idx] = [eid, year, value]
```
